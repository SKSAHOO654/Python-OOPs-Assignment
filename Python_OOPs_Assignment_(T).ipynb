5 What are dunder methods in Python ?

  - Dunder methods (short for "Double Underscore Methods") are special methods in Python that have double underscores (__) at the beginning and end of their names.

6 Explain the concept of inheritance in OOP ?

12 What is method overriding in OOP ?

Method Overriding is a concept in Object-Oriented Programming (OOP) where a child class redefines a method from its parent class with the same name and parameters to provide a different implementation.
- The method name and parameters must be the same in both parent and child classes.
- The child class method overrides the parent class method.
- Supports Runtime Polymorphism (the overridden method is determined at runtime).
- The parent method can still be accessed using super().

Both __str__ and __repr__ are special (dunder) methods in Python used to define how an object is represented as a string.

  - Used to generate a human-friendly string representation of an object.

  - Used to generate an unambiguous, developer-friendly string.
  - Called when using repr(object), in interactive mode, or object in a REPL shell.
  - Should be detailed and ideally evaluable using eval().

22 What is the difference between @staticmethod and @classmethod in Python ?

- 1. @staticmethod (Independent Utility Function)
 -  Belongs to the class, but does not access class (cls) or instance (self) variables.
 -  Works like a normal function inside a class, but logically groups it with the class.
 - Used for helper functions that do not need class-level data.

- 2. @classmethod (Class-Level Method)
 -  Operates on the class itself and takes cls as the first parameter.
 - Can modify class variables but not instance-specific data.
 -  Useful for factory methods that create instances of a class.
